```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate input vector from -10 to 10
x = np.linspace(-10, 10, 500)

# 2. Define activation functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

def leaky_relu(z, slope=0.1):
    return np.where(z > 0, z, slope * z)

# 3. Compute outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky = leaky_relu(x, slope=0.1)

# 4. Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y_sigmoid, label="Sigmoid", color="blue")
plt.plot(x, y_tanh, label="Tanh", color="red")
plt.plot(x, y_relu, label="ReLU", color="green")
plt.plot(x, y_leaky, label="Leaky ReLU (slope=0.1)", color="purple")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()


```
